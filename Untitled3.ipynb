```python
from crewai import Agent, Task, Crew
from langchain.llms import Ollama
import os
os.environ["OPENAI_API_KEY"] = "NA"
```

```python
llm1 = Ollama(
    model = "deepseek-coder-v2:16b",
    base_url = "http://localhost:11434",
    verbose = True)
```

```python
from langchain.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()
```

```python
architecht = Agent(
    role="Architect",
    goal="Based on the user input of flow charts, network architectures, mind maps etc. Write a detailed description on what are the different components, how they are connected, according to the best practices ",
    backstory="You are an expert architect that explains in detail the workflow of complex network architectures and flow charts.",
    verbose=True,
    llm=llm1,
    tools=[
        search_tool
      ]
)

coder = Agent(
  role='Coder',
  goal="""You are a helpful AI assistant who generates D2 code for diagrams based on user input. Your task is to convert the user's natural language description of a diagram into valid D2 CODE.""",
  backstory="""You are an expert coder that can write D2 code language fron any description so that the code can be used by the user to generate diagrams.""",
  verbose=True,
  llm = llm1,
  tools=[
        search_tool
      ]
)
```

```python
# Create tasks for your agents
task1 = Task(
  description="""Give d2lang code for the given cloud architecture.
  This cloud architecture utilizes a load balancer (ELB) to distribute traffic across two web servers (EC2) running a web application.
  The web servers access a relational database (RDS) for data persistence.
  A content delivery network (CDN) caches static content, reducing latency for users.
  The ELB distributes traffic to the web servers, while the web servers connect to the RDS for data access.
  The CDN serves static content directly to users, minimizing the need to access the web servers for these resources.
  This architecture ensures high availability, scalability, and improved performance for the application.""",
  agent=architecht,
  expected_output="Detailed description of cloud architecture or flow diagram",
)


task2 = Task(
  description="Based on the detailed description of the flow chart or network architecture, write correct D2 code language of the description. Additionally use the tools to write accurate D2 code",
  agent=coder,
  expected_output="D2 code corresponding to the given description, accurately representing the flowchart or network architecture."
)
```

```python
crew = Crew(
  agents=[architecht, coder],
  tasks=[task1, task2],
  verbose=2,
)
```

```python
# Get your crew to work!
result = crew.kickoff()
```
